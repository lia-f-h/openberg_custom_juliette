```python
import xarray as xr
import numpy as np
```

```python
#xr.open_mfdataset('https://thredds.met.no/thredds/catalog/accibergt5/catalog.xml')
ds = xr.open_mfdataset(['https://thredds.met.no/thredds/dodsC/accibergt5/topaz5_be_mem010.ncml',
                   'https://thredds.met.no/thredds/dodsC/accibergt5/topaz5_be_mem009.ncml'],
                         combine="nested",
    concat_dim="realization",#should be named "realization" or "ensemble_member"
    parallel=True,
    engine="netcdf4")
ds
```

```python
# create CF/OpenDrift-compliant realization coordinate
ds = ds.assign_coords(
    realization=xr.DataArray(
        ds.realization,
        dims=("realization",),
        attrs={
            "standard_name": "realization",
            "long_name": "ensemble member",
            "axis": "E"
        }
    )
)
ds
```
